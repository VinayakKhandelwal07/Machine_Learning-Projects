```python
#Kmeans Accuracy 
from sklearn.preprocessing import MinMaxScaler ,RobustScaler ,StandardScaler ,Normalizer
std = StandardScaler()

x_transformed =std.fit_transform(x)
```

```python
x_transformed
```

```python
x_transformed.shape , x.shape
```

```python
from sklearn.cluster import KMeans  # use other algos like DBSCAN , Hierachical
kmeans = KMeans()
kmeans.fit(x_transformed)
```

```python
kmeans.labels_
```

```python
np.unique(kmeans.labels_)
```

```python
kmeans.labels_.shape , df.shape
```

```python
df['cluster_no'] = kmeans.labels_
```

```python
group_six = df[df['cluster_no'] == 6]
```

```python
group_six ['label'].unique()
```

```python
import os ,joblib
os.makedirs('models',exist_ok=True)

joblib.dump(kmeans ,'./models/kmeans_model.lb')
joblib.dump(std,'./models/standardscaler.lb')
print("Your Model Has been SAved ")
```

```python
from sklearn.metrics import silhouette_score 
```

```python
silhouette_score(x_transformed,kmeans.labels_)
```
