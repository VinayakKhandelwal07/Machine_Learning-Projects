```python
# Machine Learning ML
# sklearn library will be needed for ml 
# Linear Regression

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

df = pd.read_csv('Used_Bikes.csv')
df.head()
```

```python
df.info()
```

```python
# duplicates record
print("Number of Duplicated Record : ",df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
print("Duplicates Removed!")
```

```python
print("Number of Duplicated Record : ",df.duplicated().sum())
```

```python
print("Total number of bike Record : ",df.shape[0])
```

```python
# missing value check in data
df.isnull().sum()
```

```python
# ML only deals with numerical value
df.head()
```

```python
cat_col = df.select_dtypes(include='O')
cat_col.head()
```

```python
num_col = df.select_dtypes(exclude='O')
num_col
```

```python
cat_col['owner'].value_counts()
```

```python
# Data Transformation
dt = {'First Owner':1,'Second Owner':2,'Third Owner':3,
      'Fourth Owner Or More':4}
dt
```

```python
df['owner'] = df['owner'].map(dt) # single time execute
```

```python
df
```

```python
df['brand'].value_counts()
```

```python
# Get the brands that appear at least 10 times
bike_brand = list(df['brand'].value_counts().head(16).index)
bike_brand
```

```python
df2 = df[df['brand'].isin(bike_brand)]
df2.head()
```

```python
br = {'Bajaj':1,
 'Royal Enfield':2,
 'Hero':3,
 'Honda':4,
 'Yamaha':5,
 'TVS':6,
 'KTM':7,
 'Suzuki':8,
 'Harley-Davidson':9,
 'Kawasaki':10,
 'Hyosung':11,
 'Mahindra':12,
 'Benelli':13,
 'Triumph':14,
 'Ducati':15,
 'BMW':16}

df2['brand'] = df2['brand'].map(br) # label encoding
df2.head()
```

```python
df2.drop(['bike_name','city'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
# Supervised learning   y--> price remaining x
x = df2.drop('price',axis=1) # independent variable
x
```

```python
y = df2[['price']] # dependent variable
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lnr = LinearRegression()
```

```python
lnr.fit(x_train,y_train)
```

```python
# Model Evaluation
print("Training Score : ",lnr.score(x_train,y_train)*100)
```

```python
print("Testing Score : ",lnr.score(x_test,y_test)*100)
```

```python
x_test.shape
```

```python
prediction = lnr.predict(x_test)
```

```python
prediction.shape
```

```python
comparedf = pd.concat([x_test,y_test],axis=1)
```

```python
comparedf['prediction'] = prediction
```

```python
comparedf.head(40)
```

```python
comparedf['residual'] = comparedf['price'] - comparedf['prediction']
```

```python
comparedf.head(10)
```
